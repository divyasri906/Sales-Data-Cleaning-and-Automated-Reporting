# Sales Data Cleaning and Automated Reporting

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

## Load Dataset

```python
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')

df.head()
```

## Dataset Overview

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

## Preview of Dataset

```python
df.head()
```

## Missing Values Analysis

```python
df.isnull().sum()
```

## Duplicate Records Analysis

```python
print("Duplicate Rows:", df.duplicated().sum())
```

## Data Quality Report

```python
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows:", df.duplicated().sum())

print("\nDataset is clean and ready for analysis.")
```

## Automated KPI Report

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()

print("Total Sales: $", round(total_sales, 2))
print("Total Profit: $", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
```

### Key Business Insights

- The company generated total sales of $2.29 million.
- Total profit earned was $286,397.
- A total of 5,009 orders were processed.
- The business served 793 unique customers.

## Sales by Category

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print(category_sales)
```

### Category-wise Sales Analysis

Technology generated the highest sales with approximately $836,154, followed by Furniture with $741,999 and Office Supplies with $719,047. This indicates that Technology products are the primary revenue drivers for the business.

## Category Sales Visualization

```python
plt.figure(figsize=(8,5))

category_sales.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

## Region-wise Sales Analysis

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(region_sales)
```

### Region-wise Sales Analysis

The West region generated the highest sales with approximately $725,458, followed by the East region with $678,781. The South region recorded the lowest sales among all regions. This suggests that the business has a stronger market presence in the West and East regions.

```python
plt.figure(figsize=(8,5))

region_sales.plot(kind='bar')

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

## Top 10 Customers by Sales

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

print(top_customers)
```

### Top Customers Analysis

Sean Miller was the highest revenue-generating customer with total sales of approximately $25,043. Tamara Chand and Raymond Buch were the second and third highest contributors. Identifying high-value customers helps businesses improve customer retention and targeted marketing strategies.

## Top 10 Customers by Sales Visualization

```python
plt.figure(figsize=(10,5))

top_customers.plot(kind='bar')

plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Sales')

plt.show()
```

## Automated Business Summary

```python
print("===== BUSINESS REPORT =====")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Customers: {total_customers}")

print("\nTop Category:", category_sales.idxmax())
print("Top Region:", region_sales.idxmax())
print("Top Customer:", top_customers.idxmax())
```
